```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/sample_data/fact_BusinessDetails3.csv')

# Function to convert time strings into datetime objects for easy calculation
def convert_to_time(time_str):
    if isinstance(time_str, str):  # Check if time_str is a valid string
        try:
            return datetime.strptime(time_str, "%H:%M")
        except ValueError:
            # Handle the case where the time format is invalid
            print(f"Warning: Invalid time format for {time_str}.")
            return None
    else:
        return None  # Return None for non-string values

# Function to calculate the business hours from start and end times
def calculate_work_hours(row):
    start_time = convert_to_time(row['AESTStartTime'])
    end_time = convert_to_time(row['AESTEndTime'])

    if start_time and end_time:  # Ensure both start and end times are valid
        work_hours = (end_time - start_time).seconds / 3600  # Convert seconds to hours
        return work_hours
    else:
        return 0  # If time is invalid, assume no work hours for this row

#Print result to verify data to process
print (df)
```
